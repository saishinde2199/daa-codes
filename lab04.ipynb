```python
import math as mt



x = 2
lr = 0.019        # Learning rate should be 0.01 - standard

precision = 0.000001
prev_step_size = 1
max_itr = 10000
itr = 0

gf = lambda x:  (x+3)**2

gd = []


while(precision<prev_step_size and itr<max_itr):
    prv = x
    x = x - lr * gf(prv)
    prev_step_size = abs(x-prv)

    itr += 1
    
    print("Iteration: ",itr," Value: ",x)
    gd.append(x)

```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
